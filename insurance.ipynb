```python
import pandas as pd

data = pd.read_csv("insurance.csv")

gender = data['Gender'].unique()
gender

genderT = pd.DataFrame(gender, columns=['Gender'])
genderT = genderT.reset_index()
```

```python
data
```

```python
va = data['Vehicle_Age'].unique()

va = pd.DataFrame(va, columns=['Vehicle_Age'])
va = va.reset_index()
va
```

```python
def putidforage(x):
    if x == "> 2 Years":
        age = 0
    elif x == '1-2 Year':
        age = 1
    elif x == "< 1 Year":
        age = 2
    else:
        age = "old"
    
    return age
    
data['Vehicle_Age_Id'] = data['Vehicle_Age'].apply(putidforage)

data     

```

```python
def putidforgender(x):
    if x == "Male":
        gender = 0
    elif x == 'Female':
        gender = 1
    else:
        gender = "Some Gender"
    
    return gender
    
data['Gender_Id'] = data['Gender'].apply(putidforgender)

data  
```

```python
ph = data[['id', "Gender_Id","Vehicle_Age_Id","Annual_Premium" ]]
ph
```

```python
genderT['Gender_Id'] = genderT['index']
genderT
```

```python
va['Vehicle_Age_Id'] = va['index']
```

```python
merged = pd.merge(ph, va, how="inner", on='Vehicle_Age_Id')
merged = pd.merge(merged, genderT, how='inner', on='Gender_Id')
merged = merged.drop(['index_x','index_y'], axis=1)
merged
```
