```python
import pdfplumber
import re
import csv

# Function to extract floats from text
def extract_floats(text):
    float_pattern = r'\b\d+\.\d+\b'
    return re.findall(float_pattern, text)

# Open the PDF and extract floats
pdf_path = "pdfs/1.pdf"
csv_path = "output.csv"

with pdfplumber.open(pdf_path) as pdf:
    all_floats = []
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            text = text.replace(",", "")  # Remove commas
            lines = text.split('\n')
            for line in lines:
                floats = extract_floats(line)
                if floats:
                    all_floats.append(lines)

# Write extracted floats to CSV file with each value in its column
with open(csv_path, mode="a", newline="") as file:
    writer = csv.writer(file)
    for row in all_floats:
        writer.writerow(row)

print(f"Extracted floats have been appended to {csv_path}")
```

```python
import re
import pdfplumber

def extract_text_between(start_word, end_word, text):
    # Regular expression to find text between start_word and end_word (inclusive)

    pattern = re.compile(rf'{re.escape(start_word)}\s+(.*?)\s+{re.escape(end_word)}', re.DOTALL)
    matches = pattern.findall(text)
    return matches

# Open the PDF
with pdfplumber.open("pdfs/15.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text()  # Extract text from the page
        text = text.replace(",", "")  # Remove commas for consistent parsing
        
        sx = text.split('\n')
        for line in sx:
            # Example: looking for text starting with 'start_word' and ending with 'end_word'
            start_word = "H/Q"
            end_word = "Nommer"
            
            matches = extract_text_between(start_word, end_word, line)
            if matches:
                for match in matches:
                    print(f"Extracted text: {match}")
```

```python
import pdfplumber
import re

# Function to extract amounts (e.g., $1,234.56, 1234.56)
def extract_amounts(text):
    # Regular expression to capture amounts, including currency symbols, commas, and decimals
    amount_pattern = r'\b(?:\$|€|£|¥)?\d{1,3}(?:,\d{3})*(?:\.\d{2})?\b'
    return re.findall(amount_pattern, text)

# Open the PDF
with pdfplumber.open("pdfs/3.pdf") as pdf:
    all_amounts = []
    for page in pdf.pages:
        text = page.extract_text()  # Extract text from the page
        if text:
            amounts = extract_amounts(text)  # Extract amounts from the text
            all_amounts.extend(amounts)

    # Optional: Clean up and convert amounts (e.g., remove currency symbols, convert commas)
    cleaned_amounts = []
    for amount in all_amounts:
        cleaned = amount.replace(',', '')  # Remove commas for numeric conversion
        cleaned_amounts.append(cleaned)

    print(cleaned_amounts)
```
